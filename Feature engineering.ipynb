```python
import pandas as pd
import numpy as np
import sklearn
import plotly.express as px
import plotly.graph_objects as go
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import QuantileTransformer
from sklearn.ensemble import RandomForestClassifier
```

```python
#df_train_no_nan=pd.read_csv('df_train_no_Nan.csv')
df_train_balanced=pd.read_csv('df_train_balanced.csv')
```

```python
#df_numerical_nan=df_train_no_nan.select_dtypes(['float','int'])
df_numerical_balanced=df_train_balanced.select_dtypes(['float','int'])
```

```python
qt = QuantileTransformer(output_distribution='normal')
```

```python
#ls_all_nan=df_numerical_nan.columns.tolist()
ls_all_balanced=df_numerical_balanced.columns.tolist()
```

```python
#del ls_all_nan[1]
del ls_all_balanced[1]
```

```python
for k in ls_all_nan:
    df_train_no_nan[k] = qt.fit_transform(df_train_no_nan[[k]].values.reshape(-1,1))
```

```python
for k in ls_all_balanced:
    df_train_balanced[k] = qt.fit_transform(df_train_balanced[[k]].values.reshape(-1,1))
```

```python
df_train_balanced
```

```python
#categorical_variables_nan=df_train_no_nan.select_dtypes(exclude=['int','float'])
categorical_variables_balanced=df_train_balanced.select_dtypes(exclude=['int','float'])
```

```python
#ls_nan=categorical_variables_nan.columns.tolist()
ls_balanced=categorical_variables_balanced.columns.tolist()
```

```python
df_dummies_nan=pd.get_dummies(data=df_train_no_nan, columns=ls_nan)
df_dummies_nan
```

```python
df_dummies_balanced=pd.get_dummies(data=df_train_balanced, columns=ls_balanced)
df_dummies_balanced
```

```python
df_dummies_balanced.to_csv('df_balanced.csv')
```
