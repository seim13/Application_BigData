# Feature engineering

```python
import pandas as pd
import numpy as np
import sklearn

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import QuantileTransformer
from sklearn.ensemble import RandomForestClassifier
```

```python
df_train_balanced=pd.read_csv('df_train_balanced.csv')
```

```python
df_test=pd.read_csv('df_test.csv')
```

```python
df_train_balanced
```

```python
df_train_balanced.dtypes
```

## Normalize the rest numerical column

```python
df_numerical_balanced=df_train_balanced.select_dtypes(['float','int'])
df_numerical_test=df_test.select_dtypes(['float','int'])
```

```python
df_numerical_balanced
```

```python
qt = QuantileTransformer(output_distribution='normal')
```

```python
ls_all_balanced=df_numerical_balanced.columns.tolist()
ls_test=df_numerical_test.columns.tolist()
```

```python
ls_all_balanced
```

```python
del ls_all_balanced[1]
```

```python
for k in ls_all_balanced:
    df_train_balanced[k] = qt.fit_transform(df_train_balanced[[k]].values.reshape(-1,1))
```

```python
for k in ls_test:
    df_test[k] = qt.fit_transform(df_test[[k]].values.reshape(-1,1))
```

```python
df_train_balanced
```

## One encode categoriacl data 

```python
categorical_variables_balanced=df_train_balanced.select_dtypes(exclude=['int','float'])
categorical_variables_test=df_test.select_dtypes(exclude=['int','float'])
```

```python
ls_balanced=categorical_variables_balanced.columns.tolist()
ls_test=categorical_variables_test.columns.tolist()
```

```python
ls_balanced
```

```python
df_dummies_balanced=pd.get_dummies(data=df_train_balanced, columns=ls_balanced)
df_dummies_balanced
```

```python
df_dummies_test=pd.get_dummies(data=df_test, columns=ls_test)
df_dummies_test
```

## Saving new dataset

```python
df_dummies_test.to_csv('df_clean_test.csv',index=False)
```

```python
df_dummies_balanced.to_csv('df_balanced.csv',index=False)
```
